```python
# Import the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Dataset.csv')
```

```python
df = data.dropna()
```

```python
df.columns
```

```python
pd.crosstab(df.Target,df["Failure Type"])
```

```python
df["Rotational speed [rpm]"].hist(figsize=(7,4), color = 'red')
plt.title("Rotational distribution")
plt.xlabel("Rotational speed")
plt.ylabel("Speed Calculations")
plt.show()
```

```python
df["Process temperature [K]"].plot(figsize=(7,4), color='green')
plt.title("Temperature distribution")
plt.xlabel("Temperature in Kelvin")
plt.ylabel("No of Counts")
plt.show()
```

```python
# Heatmap plot diagram
fig, ax = plt.subplots(figsize=(15,6))
sns.heatmap(df, ax=ax, annot=True)
```

```python
def Contract(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(6,6), autopct='%1.2f%%', fontsize = 10)
    ax.set_title(variable + ' \n', fontsize = 10)
    
Contract(df, 'Type')
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(x=df['Air temperature [K]'],y=df["Process temperature [K]"])
plt.show()
```
